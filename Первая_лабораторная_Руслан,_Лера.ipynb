Кодирование категориальных данных в столбце Target. Так как цели данных сложно связать и лишние зависимости нам не нужны, то применяем OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder

# Создаем объект OneHotEncoder
enc = OneHotEncoder(sparse=False)

res = enc.fit_transform(students[['Target']])
enc.categories_
enc.categories_[0]
students[enc.categories_[0]] = res
students.head()
```

Визуализация данных датафрейма для признаков

```python
# Строим гистограмму для каждого признака
for col in students.columns:
  plt.hist(students[col])
  plt.title(col)
  plt.show()
```

```python
# Строим круговые диаграммы для каждого признака
for column in students.columns:
  # Подсчитываем количество значений для каждой категории
  value_counts = students[column].value_counts()

  # Строим круговую диаграмму
  plt.pie(value_counts.values, labels=value_counts.index)
  plt.title(column)
  plt.show()
```

```python
# Выбираем числовые признаки
numeric_columns = students.select_dtypes(include=['number']).columns

# Строим ящики с усами для каждого признака
for column in numeric_columns:
  sns.boxplot(x=students[column])
  plt.title(column)
  plt.show()
```
